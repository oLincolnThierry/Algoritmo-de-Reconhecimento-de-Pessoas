```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/M-02.jfif")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # HAAR cascade funciona melhor em escalas de cinza do que em imagens com cor.
```

```python
## Carregando o modelo HAAR Cascade [XML]
person_classifier_file = "/haarcascade_fullbody.xml"
person_cascade = cv2.CascadeClassifier(person_classifier_file) # metodo classificador em cascata
## Gerando previsão
persons = person_cascade.detectMultiScale(gray_img)
```

```python
for (x,y,w,h) in persons:
  gray = cv2.rectangle(img,(x,y),(x+w,y+h),(70,0,0),2) #define que o retangulo vai ser posicionado no canto SP(x+w) ID(y+h), RGB e Espessura do retangulo
  font = cv2.FONT_HERSHEY_COMPLEX_SMALL
  cv2.putText(img,'pessoa',(x,y -7), font, 1.0,(11,20,180), 1) # Adiciona Label nos retangulos dizendo que são pessoas
```
